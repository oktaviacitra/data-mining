# Read Dataset

```python
path = "/content/drive/MyDrive/Data Mining/transaction.csv"
dataset = pd.read_csv(path)
dataset
```

```python
data = dataset.loc[dataset['Country'] == 'Portugal']
data
```

```python
transaction = data.groupby(['InvoiceNo', 'StockCode'])['Qty'].sum()
transaction = transaction.unstack().reset_index().fillna(0).set_index('InvoiceNo')
transaction[transaction > 0] = 1
transaction.columns = transaction.columns.map(str)
transaction
```

```python
frequent_itemsets = apriori(transaction, min_support=0.2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules[['antecedents', 'consequents', 'confidence']]
```
